### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import statistics

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

def freqyency (c):
    b = sorted(c)
    d = {}
    for i in b:
        if i in d:
            d[i]= d[i]+1
        else:
            d[i] = 1
    return d

print ('Максимальне значення ', max(list_with_random_numbers))
print ('Мінімальне значення ', min(list_with_random_numbers))
print ('Середнє значення послідовності ',statistics.mean(list_with_random_numbers))
print ('Медіана послідовності ',statistics.median(list_with_random_numbers))
print ('Частота кожної цифри в послідовності ', freqyency (list_with_random_numbers))
print ('Середнє квадратичне відхилення ', statistics.stdev(list_with_random_numbers))




                    
```
